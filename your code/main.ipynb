# Bonus Challenge 1 - Making Your Spider Unblockable

Use techniques such as randomizing user agents and referers in your requests to reduce the likelihood that your spider is blocked by websites. [Here](http://blog.adnansiddiqi.me/5-strategies-to-write-unblock-able-web-scrapers-in-python/) is a great article to learn these techniques.

In the cell below, place your entire code including the updated `IronhackSpdier` class and the code to kickstart the spider.

```python
import time
import requests
from bs4 import BeautifulSoup

class IronhackSpider:

    def __init__(self, url_pattern, pages_to_scrape=10, sleep_interval=-1, content_parser=None):
        self.url_pattern = url_pattern
        self.pages_to_scrape = pages_to_scrape
        self.sleep_interval = sleep_interval
        self.content_parser = content_parser

    def scrape_url(self, url):
        headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36',
            'Referer': 'https://www.google.com/'
        }

        try:
            response = requests.get(url, headers=headers)
            response.raise_for_status()
            result = self.content_parser(response.content)
            self.output_results(result)

        except requests.exceptions.HTTPError as http_err:
            print(f"HTTP error occurred while scraping {url}: {http_err}")
        except requests.exceptions.ConnectionError as conn_err:
            print(f"Connection error occurred while scraping {url}: {conn_err}")
        except requests.exceptions.Timeout as timeout_err:
            print(f"Timeout error occurred while scraping {url}: {timeout_err}")
        except Exception as err:
            print(f"An unexpected error occurred while scraping {url}: {err}")

    def output_results(self, r):
        print(r)

    def kickstart(self):
        for i in range(1, self.pages_to_scrape + 1):
            self.scrape_url(self.url_pattern % i)

            # Sleep if interval is set and not the last page
            if self.sleep_interval > 0 and i != self.pages_to_scrape:
                print(f"Sleeping for {self.sleep_interval} seconds before next request...")
                time.sleep(self.sleep_interval)

# Example use case
URL_PATTERN = 'https://books.toscrape.com/catalogue/page-%s.html'
PAGES_TO_SCRAPE = 5

def books_parser(content):
    soup = BeautifulSoup(content, 'html.parser')
    books = []

    for book_block in soup.select(".product_pod"):
        title = book_block.h3.a['title']
        books.append({"title": title})

    return books

# Instantiate and run spider with a delay between pages
my_spider = IronhackSpider(
    url_pattern=URL_PATTERN,
    pages_to_scrape=PAGES_TO_SCRAPE,
    sleep_interval=2,  # delay in seconds
    content_parser=books_parser
)

my_spider.kickstart()
```

```python
import aiohttp
import asyncio
import nest_asyncio  # Hier importieren wir nest_asyncio
import time
from bs4 import BeautifulSoup

# Anwenden von nest_asyncio, um den bestehenden Event-Loop zu ermöglichen
nest_asyncio.apply()

class IronhackSpider:

    def __init__(self, url_pattern, pages_to_scrape=10, sleep_interval=-1, content_parser=None):
        self.url_pattern = url_pattern
        self.pages_to_scrape = pages_to_scrape
        self.sleep_interval = sleep_interval
        self.content_parser = content_parser
        self.headers = {
            'User-Agent': 'Mozilla/5.0 (...)',  # Dein User-Agent hier
            'Referer': 'https://www.google.com/'
        }

    async def scrape_url(self, session, url):
        try:
            async with session.get(url, headers=self.headers) as response:
                response.raise_for_status()
                content = await response.text()
                result = self.content_parser(content)
                self.output_results(result)
        
        except Exception as e:
            print(f"Error while scraping {url}: {e}")

    def output_results(self, r):
        print(r)

    async def scrape_pages(self):
        async with aiohttp.ClientSession() as session:
            tasks = []
            for i in range(1, self.pages_to_scrape + 1):
                url = self.url_pattern % i
                task = asyncio.create_task(self.scrape_url(session, url))
                tasks.append(task)

                # Adding sleep interval between requests, if needed
                if self.sleep_interval > 0:
                    await asyncio.sleep(self.sleep_interval)
            
            await asyncio.gather(*tasks)

    def kickstart(self):
        # Direkter Start des Event-Loops
        asyncio.get_event_loop().run_until_complete(self.scrape_pages())

URL_PATTERN = 'https://books.toscrape.com/catalogue/page-%s.html'
PAGES_TO_SCRAPE = 5

def books_parser(content):
    soup = BeautifulSoup(content, 'html.parser')
    books = []
    for book_block in soup.select(".product_pod"):
        title = book_block.h3.a['title']
        books.append({"title": title})
    return books

# Instantiate the IronhackSpider class
my_spider = IronhackSpider(URL_PATTERN, PAGES_TO_SCRAPE, content_parser=books_parser)

# Start scraping jobs
my_spider.kickstart()
```
